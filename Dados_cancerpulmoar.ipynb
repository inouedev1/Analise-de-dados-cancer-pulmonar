```python
import json
```

```python
import pandas as pd
```

```python
path = "/content/cancer patient data sets.csv"
```

```python
df = pd.read_csv(path)
```

```python
display(df)
```

```python
df.columns
```

```python
df = df.drop(columns=['Patient Id', 'OccuPational Hazards','Coughing of Blood', 'Fatigue', 'Clubbing of Finger Nails' ,'Swallowing Difficulty', 'Shortness of Breath','Snoring'] )
```

```python
df['Gender'] = df['Gender'].replace({1: 'M', 2: 'F'})
```

```python
#df
```

```python
import plotly.express as px
grafico_exposicao_ar = px.histogram(df, x='Air Pollution',
nbins=20, title='Numero de casos por exposição a poluição do ar',
labels={'Air Pollution': 'Nivel De Exposição', 'count': 'Número de Casos'})
grafico_exposicao_ar
```

```python
grafico_idade_casos = px.histogram(df, x='Age', nbins=20,
title='Idade Distribuída por frequencia de casos',
labels={'Age': 'Idade', 'count': 'Número de Casos'})
grafico_idade_casos
```

```python
grafico_alergia_casos = px.histogram(df, x='Dust Allergy',
nbins=20, title='Alergia a pó',
labels={'Dust Allergy': 'Nível de alergia', 'count': 'Número de Casos'})
grafico_alergia_casos
```

```python
genero_count = df['Gender'].value_counts().reset_index()
genero_count.columns = ['Gênero', 'Número de Pacientes']
grafico_genero_casos = px.bar(genero_count, x='Gênero', y='Número de Pacientes', title='Frequência de Pessoas com Câncer por Gênero')
grafico_genero_casos
```

```python
grafico_alcool_casos = px.histogram(df, x='Alcohol use',
nbins=20, title='Uso de alcool',
labels={'Alcohol use': 'Nível de Uso de Álcool', 'count': 'Frequência'})
grafico_alcool_casos
```

```python
grafico_fumantesa_casos = px.histogram(df, x='Smoking', nbins=20,
title='Fumantes ativos',
labels={'Smoking': 'Nível de Fumantes', 'count': 'Frequência'})
grafico_fumantesa_casos
```

```python
grafico_fumantesp_casos = px.histogram(df, x='Passive Smoker', nbins=20,
title='Fumantes Passivos',
labels={'Passive Smoker': 'Nível de Fumantes', 'count': 'Frequência'})
grafico_fumantesp_casos
```

```python
grafico_analise1_compar = px.histogram(df, x=['Smoking', 'Passive Smoker'], nbins=20,
title='Comparação entre Fumantes Ativos e Fumantes Passivos',
labels={'value': 'Nível de Fumantes', 'count': 'Frequência'}, marginal='rug')
grafico_analise1_compar
```

```python
grafico_analise2_compar = px.histogram(df, x=['Smoking', 'Passive Smoker', 'Air Pollution', 'Alcohol use', 'Dust Allergy'], nbins=20,
title='Comparação entre os principais Fatores de Risco para Câncer',
labels={'value': 'Nível', 'count': 'Frequência'}, marginal='rug')
grafico_analise2_compar
```

```python
fig_poluicao_alergia = px.histogram(df, x=['Air Pollution', 'Dust Allergy'], nbins=20,
title='Comparação entre Poluição do Ar e Alergia a Pó',
labels={'value': 'Nível', 'count': 'Frequência'}, marginal='rug')
fig_poluicao_alergia
```
